```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

# QE Financial Spillover Dynamics

```python
data = pd.read_csv('data_t.csv')
```

## Overview

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12,8))

ax1.plot(data['FX_domestic_foreign'].index, data['FX_domestic_foreign'])
ax1.set_ylabel('FX domestic foreign', fontsize='12')

ax2.plot(data['asset0price'].index, data['asset0price'], label='asset0')
ax2.plot(data['asset1price'].index, data['asset1price'], label='asset1')
ax2.set_ylabel('asset prices', fontsize='12')
ax2.legend(frameon=True, loc='best', framealpha=0.8)

ax3.plot(data['redeemable_shares_fund_0'].index, data['redeemable_shares_fund_0'], label='fund0')
ax3.plot(data['redeemable_shares_fund_1'].index, data['redeemable_shares_fund_1'], label='fund1')
ax3.set_ylabel('Redeemable shares', fontsize='12')
ax3.legend(frameon=True, loc='best', framealpha=0.8)

ax4.plot(data['Delta_asset0'].index, data['Delta_asset0'], label='asset0')
ax4.plot(data['Delta_asset1'].index, data['Delta_asset1'], label='asset1')
#ax4.plot(data['Delta_FX'].index, data['Delta_FX'], label='fx')
ax4.set_ylabel('Delta', fontsize='12')
ax4.legend(frameon=True, loc='best', framealpha=0.8)
```

# Expectations

## Expected default rates

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,8))

ax1.plot(data['exp_default_asset0_fund_0'].index, data['exp_default_asset0_fund_0'], label='fund0')
ax1.plot(data['exp_default_asset0_fund_1'].index, data['exp_default_asset0_fund_1'], label='fund1')
ax1.set_ylabel('exp_default_asset0', fontsize='12')
ax1.legend(frameon=True, loc='best', framealpha=0.8)

ax2.plot(data['exp_default_asset1_fund_0'].index, data['exp_default_asset1_fund_0'], label='fund0')
ax2.plot(data['exp_default_asset1_fund_1'].index, data['exp_default_asset1_fund_1'], label='fund1')
ax2.set_ylabel('exp_default_asset1', fontsize='12')
ax2.legend(frameon=True, loc='best', framealpha=0.8)
```

## Expected returns

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12,8))

ax1.plot(data['exp_return_asset0_fund_0'].index, data['exp_return_asset0_fund_0'], label='fund0')
ax1.plot(data['exp_return_asset0_fund_1'].index, data['exp_return_asset0_fund_1'], label='fund1')
ax1.set_ylabel('exp return asset 0', fontsize='12')
ax1.legend(frameon=True, loc='best', framealpha=0.8)

ax2.plot(data['exp_return_asset1_fund_0'].index, data['exp_return_asset1_fund_0'], label='fund0')
ax2.plot(data['exp_return_asset1_fund_1'].index, data['exp_return_asset1_fund_1'], label='fund1')
ax2.set_ylabel('exp return asset 1', fontsize='12')
ax2.legend(frameon=True, loc='best', framealpha=0.8)

ax3.plot(data['exp_return_currency0_fund_0'].index, data['exp_return_currency0_fund_0'], label='fund0')
ax3.plot(data['exp_return_currency0_fund_1'].index, data['exp_return_currency0_fund_1'], label='fund1')
ax3.set_ylabel('exp return currency 0', fontsize='12')
ax3.legend(frameon=True, loc='best', framealpha=0.8)

ax4.plot(data['exp_return_currency1_fund_0'].index, data['exp_return_currency1_fund_0'], label='fund0')
ax4.plot(data['exp_return_currency1_fund_1'].index, data['exp_return_currency1_fund_1'], label='fund1')
ax4.set_ylabel('exp return currency 1', fontsize='12')
ax4.legend(frameon=True, loc='best', framealpha=0.8)
```

# Quantities

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4))

ax1.plot(data['a_quantityasset0_fund_0'].index, data['a_quantityasset0_fund_0'], label='asset0')
ax1.plot(data['a_quantityasset1_fund_0'].index, data['a_quantityasset1_fund_0'], label='asset1')
ax1.plot(data['c_quantitycurrency0_fund_0'].index, data['c_quantitycurrency0_fund_0'], label='currency0')
ax1.plot(data['c_quantitycurrency1_fund_0'].index, data['c_quantitycurrency1_fund_0'], label='currency1')
ax1.set_ylabel('fund0', fontsize='12')
ax1.legend(frameon=True, loc='best', framealpha=0.8)

ax2.plot(data['a_quantityasset0_fund_1'].index, data['a_quantityasset0_fund_1'], label='asset0')
ax2.plot(data['a_quantityasset1_fund_1'].index, data['a_quantityasset1_fund_1'], label='asset1')
ax2.plot(data['c_quantitycurrency0_fund_1'].index, data['c_quantitycurrency0_fund_1'], label='currency0')
ax2.plot(data['c_quantitycurrency1_fund_1'].index, data['c_quantitycurrency1_fund_1'], label='currency1')

ax2.set_ylabel('fund1', fontsize='12')
ax2.legend(frameon=True, loc='best', framealpha=0.8)
```
